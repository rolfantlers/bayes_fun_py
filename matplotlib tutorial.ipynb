```python
# creates a random normal walk and plots
np.random.seed(42) # why is this called?
y = np.random.standard_normal(50)
x = np.arange(len(y))
plt.plot(x,y);
```

```python
y1 = np.arange(100)
y1 = np.cumsum(y1)
x1 = np.arange(len(y1))
plt.plot(x1,y1)
```
